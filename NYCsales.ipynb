```python
#change column name
nycc = nyc.rename(columns={'ZIP CODE': 'ZIP','RESIDENTIAL UNITS': 'RESIDENTIAL', 'COMMERCIAL UNITS': 'COMMERCIAL', 
                           'TOTAL UNITS': 'TOTAL','YEAR BUILT': 'YEARB', 'TAX CLASS AT TIME OF SALE': 'TAXCLSA'})
```
